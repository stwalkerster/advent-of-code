```python
inputExample = open('day24-example.txt', 'r').read().split('\n')
inputReal = open('day24.txt', 'r').read().split('\n')
```

```python
def parse(input):
    result = []
    for line in input:
        chars = list(line)
        instructions = []
        while len(chars) > 0:
            c = chars.pop(0)

            if c in ['s', 'n']:
                c += chars.pop(0)

            instructions.append(c)
        result.append(instructions)
    return result
```

```python
def part1(input):
    instructions = parse(input)

    blackTiles = []

    directions = {
        # q, r, s
        'e':  (1, 0, -1),
        'se': (0, 1, -1),
        'sw': (-1, 1, 0),
        'w':  (-1, 0, 1),
        'nw': (0, -1, 1),
        'ne': (1, -1, 0)
    }

    for line in instructions:
        tile = (0,0,0)
        for i in line:
            tile = vectorAdd(tile, directions[i])

        if tile in blackTiles:
            blackTiles.remove(tile)
        else:
            blackTiles.append(tile)

    print(len(blackTiles))
```

```python
def vectorAdd(a, b):
    return a[0] + b[0], a[1] + b[1], a[2] + b[2]
```

```python
part1(inputExample)
```

```python
part1(inputReal)
```

```python
def part2(input):
    instructions = parse(input)

    blackTiles = []

    directions = {
        # q, r, s
        'e':  (1, 0, -1),
        'se': (0, 1, -1),
        'sw': (-1, 1, 0),
        'w':  (-1, 0, 1),
        'nw': (0, -1, 1),
        'ne': (1, -1, 0)
    }

    # load tiles
    for line in instructions:
        tile = (0,0,0)
        for i in line:
            tile = vectorAdd(tile, directions[i])

        if tile in blackTiles:
            blackTiles.remove(tile)
        else:
            blackTiles.append(tile)

    print('Day', 0, ':', len(blackTiles))
    # print(blackTiles)

    def countSurrounding(tile):
        count = 0
        for d in directions:
            if vectorAdd(tile, directions[d]) in blackTiles:
                count += 1
        return count

    # emulate flips
    for day in range(100):
        newBlacks = []
        for b in blackTiles:
            neighbours = countSurrounding(b)
            #print("  tile", b, "has", neighbours, "neighbours")
            if neighbours == 0 or neighbours > 2:
                pass
            else:
                newBlacks.append(b)

            for d in directions:
                w = vectorAdd(b, directions[d])
                if w not in blackTiles:
                    neighbours = countSurrounding(w)
                    if neighbours == 2 and w not in newBlacks:
                        newBlacks.append(w)

        print("Day", day+1, ":", len(newBlacks))
        blackTiles = newBlacks

```

```python
part2(inputExample)
```
