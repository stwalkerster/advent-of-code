```python
inputExample = open('example.txt', 'r').read().split('\n')
inputReal = open('input.txt', 'r').read().split('\n')
```

```python
def parse(input):
    pairs = []

    for line in input:
        # print(line)
        splits = line.split(' -> ')
        for i in range(1, len(splits)):
            pair1 = [int(i) for i in splits[i].split(',')]
            pair2 = [int(i) for i in splits[i-1].split(',')]

            pairs.append([pair2, pair1])

    return pairs
```

```python
def part1(input):
    pairs = parse(input)

    # find grid bounds
    minY = min(0, *[min([i[1] for i in p]) for p in pairs])
    maxY = max(0, *[max([i[1] for i in p]) for p in pairs])
    minX = min(500, *[min([i[0] for i in p]) for p in pairs])
    maxX = max(500, *[max([i[0] for i in p]) for p in pairs])

    deltaY = 0 - minY
    deltaX = 0 - minX + 1

    input = [500 + deltaX, 0 + deltaY]
    grid = [['.' for x in range(minX-1, maxX+2)] for y in range(minY,maxY+2)]
    grid[input[1]][input[0]] = '+'

    for p in pairs:
        x_values = [p[0][0], p[1][0]]
        x_values.sort()
        for x in range(x_values[0], x_values[1] + 1):
            y_values = [p[0][1], p[1][1]]
            y_values.sort()
            for y in range(y_values[0], y_values[1]+1):
                grid[y + deltaY][x + deltaX] = '#'

    # print_grid(grid)

    i = 0
    while True:
        # print()
        #print("Loop", i, input)
        grain = input[::]
        moved = True
        while moved:
            moved = False

            # fallen off the grid
            if grain[1] == (len(grid)-1):
                #print("inner break")
                break

            if grid[grain[1]+1][grain[0]] == '.':
                grain[1] += 1
                moved = True
                #print ("down")
                continue

            if grid[grain[1]+1][grain[0]-1] == '.':
                grain[1] += 1
                grain[0] -= 1
                moved = True
                #print ("left")
                continue

            if grid[grain[1]+1][grain[0]+1] == '.':
                grain[1] += 1
                grain[0] += 1
                moved = True
                #print ("right")
                continue

        # fallen off the grid as it's on the bottom row
        if grain[1] == (len(grid)-1):
            print_grid(grid)
            print("Void hit at i =", i)
            break

        #print("positioning grain at", grain)
        grid[grain[1]][grain[0]] = 'o'
        i += 1
```

```python
def print_grid(grid, file=None):
    x_limit = len(grid[0])
    if file is None:
        x_limit = min(x_limit, 100)

    for y in range(len(grid)):
        for x in range(x_limit):
            print(grid[y][x], end='',file=file)
        print(file=file)
```

```python
def part2(input):
    pairs = parse(input)

    # find grid bounds
    minY = min(0, *[min([i[1] for i in p]) for p in pairs])
    maxY = max(0, *[max([i[1] for i in p]) for p in pairs])
    minX = min(500, *[min([i[0] for i in p]) for p in pairs])
    maxX = max(500, *[max([i[0] for i in p]) for p in pairs])

    deltaY = 0
    deltaX = 0

    input = [500 + deltaX, 0 + deltaY]
    grid = [['.' for x in range(0, 1000)] for y in range(0,maxY + 2)]
    grid[input[1]][input[0]] = '+'

    for p in pairs:
        x_values = [p[0][0], p[1][0]]
        x_values.sort()
        for x in range(x_values[0], x_values[1] + 1):
            y_values = [p[0][1], p[1][1]]
            y_values.sort()
            for y in range(y_values[0], y_values[1]+1):
                grid[y + deltaY][x + deltaX] = '#'

    # print_grid(grid)

    i = 0
    while True:
        # print()
        # print("Loop", i, input)
        grain = input[::]
        moved = True
        while moved:
            moved = False

            # fallen off the grid
            if grain[1] == (len(grid)-1):
                # print("inner break")
                break

            if grid[grain[1]+1][grain[0]] == '.':
                grain[1] += 1
                moved = True
                # print ("down")
                continue

            if grid[grain[1]+1][grain[0]-1] == '.':
                grain[1] += 1
                grain[0] -= 1
                moved = True
                # print ("left")
                continue

            if grid[grain[1]+1][grain[0]+1] == '.':
                grain[1] += 1
                grain[0] += 1
                moved = True
                # print ("right")
                continue

        # print("positioning grain at", grain)
        grid[grain[1]][grain[0]] = 'o'

        if i % 1000 == 0:
            f = open('output.txt', 'w')
            print_grid(grid, f)
            f.close()

        i += 1

        if grain == input:
            print("end at i=", i)

            f = open('output.txt', 'w')
            print_grid(grid, f)
            f.close()
            break

```
